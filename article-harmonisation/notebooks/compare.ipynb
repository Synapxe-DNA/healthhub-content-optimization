## Import Relevant Libraries

```python
import pandas as pd
```

## Load Older Dataframe

```python
relevant_categories = [
    "cost-and-financing",
    "live-healthy-articles",
    "diseases-and-conditions",
    "medical-care-and-facilities",
    "support-group-and-others",
]
```

```python
# Load dataframe from `data` directory
older_df = pd.read_parquet("../data/merged_data_older.parquet")
older_df = older_df.drop(columns=["related_sections", "extracted_images", "extracted_tables", "extracted_links", "extracted_headers"])

display(older_df)
```

## Filter by Provider Name, Relevant Content Categories, Those to Keep

```python
older_df_hpb = older_df[(older_df["pr_name"] == "Health Promotion Board")]

display(older_df_hpb)  # noqa: F821
```

```python
print(older_df_hpb["content_category"].value_counts())
```

```python
older_df_hpb = older_df_hpb[(older_df_hpb["content_category"].isin(relevant_categories))]

display(older_df_hpb)  # noqa: F821
```

```python
print(older_df_hpb["content_category"].value_counts())
```

```python
older_df_hpb = older_df_hpb[(~older_df_hpb["to_remove"])]

display(older_df_hpb)  # noqa: F821
```
