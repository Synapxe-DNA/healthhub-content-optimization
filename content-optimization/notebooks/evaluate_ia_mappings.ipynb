# Filter Data

Focus on the 5 key content categories and selected columns of interest

```python
df = merged_data[
    [
        "id",
        "content_name",
        "title",
        "full_url",
        "content_body",
        "extracted_content_body",
        "pr_name",
        "date_modified",
        "content_category",
        "article_category_names",
        "l1_mappings",
        "l2_mappings",
    ]
]

df = df[
    df["content_category"].isin(
        [
            "cost-and-financing",
            "diseases-and-conditions",
            "live-healthy-articles",
            "medical-care-and-facilities",
            "support-group-and-others",
        ]
    )
]

display(df)
```

## Find Rows with Null L1 and L2 Mappings

```python
null_df = df[(df["l1_mappings"].isnull()) & (df["l2_mappings"].isnull())]

display(null_df)
```

```python
null_df_counts = (
    null_df.groupby(by=["content_category", "article_category_names"], dropna=False)
    .size()
    .sort_values(ascending=False)
)

null_df_counts
```

## Extract Combinations for Null L1 & L2 Mappings (Missing)

```python
null_mappings = list(null_df_counts.keys())
```

```python
for categories in null_mappings:
    print(categories)
```

```python
def extract_unqiue_categories_dict(categories):
    result = {}
    for content_category, article_category_names in categories:
        if content_category not in result:
            result[content_category] = set()

        # Skip non-string values (e.g. nan)
        if not isinstance(article_category_names, str):
            print(content_category, article_category_names)
            continue

        article_categories = article_category_names.split(",")
        for article_category in article_categories:
            cleaned_category = article_category.strip()
            if cleaned_category != "":
                result[content_category].add(cleaned_category)

    for key, value in result.items():
        result[key] = sorted(list(value))

    return result


category_mappings = extract_unqiue_categories_dict(null_mappings)
print(category_mappings)
```

## Extract values counts given that some articles can have multiple sub-categories

```python
for key, values in category_mappings.items():
    for sub_category in values:
        null_df_slice = null_df[
            (null_df["content_category"] == key)
            & (null_df["article_category_names"].str.contains(sub_category))
        ]
        print(f"Value Counts for {key}, {sub_category}:", null_df_slice.shape[0])
```

```python
for key in category_mappings.keys():
    null_df_slice = null_df[
        (null_df["content_category"] == key)
        & (null_df["article_category_names"].isnull())
    ]
    print(f"Value Counts for {key}, null:", null_df_slice.shape[0])
```
