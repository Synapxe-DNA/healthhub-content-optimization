```python
# ruff: noqa: F821
merged_data = catalog.load("merged_data")

display(merged_data)
```

```python
cols = [
    "id",
    "content_name",
    "full_url",
    "friendly_url",
    "title",
    "content_category",
    "content_body",
    "extracted_content_body",
]

df = merged_data[cols]
```

```python
def find_chinese(text):
    if text is None:
        return False
    if result := re.search("[\u4e00-\u9FFF]", text):
        print(result.group(0))
        return True
    else:
        return False


df["chinese_text"] = df["content_body"].apply(find_chinese)
```

```python
df_chinese = df[df["chinese_text"]]

display(df_chinese)
```

```python
def find_tamil(text):
    if text is None:
        return False
    if result := re.search("[\u0B80-\u0BFF]", text):
        print(result.group(0))
        return True
    else:
        return False


df["tamil_text"] = df["content_body"].apply(find_tamil)
```

```python
df_tamil = df[df["tamil_text"]]

display(df_tamil)
```

```python
def find_multilingual(friendly_url):
    if friendly_url is None:
        return False

    check_lang = friendly_url.split("_")[-1]
    check_lang = check_lang.split("-")[-1]
    if result := re.search("(chinese|tamil|malay)", check_lang):
        print(result.group(0))
        return True
    else:
        return False


df["not_english"] = df["friendly_url"].apply(find_multilingual)
```

```python
df_multilingual = df[df["not_english"]].sort_values(by=["friendly_url"])

display(df_multilingual)
```
