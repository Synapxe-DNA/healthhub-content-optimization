```python
# ruff: noqa: F821
merged_data = catalog.load("merged_data")
merged_data
```

```python
# Remove NaN extracted content body
tmp = merged_data.query("extracted_content_body.notna()")
# Calculate the number of words in the extracted content body
tmp["word_count"] = tmp["extracted_content_body"].str.split().str.len()
# Remove all words less than 1
tmp = tmp[tmp["word_count"] > 0].reset_index(drop=True)

tmp.shape
```

```python
tmp["log_word_count"] = np.log(tmp["word_count"])
```

```python
fig = px.histogram(tmp, x="word_count", nbins=100)
fig.update_layout(
    title_text="Raw word count distribution",
    xaxis_title_text="Word count",
    yaxis_title_text="Count",
    bargap=0.1,
)
fig.add_vline(x=4.5, line_dash="dash", line_color="firebrick")
fig.show()
```

```python
# # ruff: noqa: F821
catalog.save("raw_word_counts", fig)
```

```python
fig = px.histogram(tmp, x="log_word_count", nbins=100)
fig.update_layout(
    title_text="log(word_count) distribution: np.exp(4.5) ≈ 90",
    xaxis_title_text="log(word_count)",
    yaxis_title_text="Count",
    bargap=0.1,
)
fig.add_vline(x=4.5, line_dash="dash", line_color="firebrick")
fig.show()
```

```python
# # ruff: noqa: F821
catalog.save("log_word_counts", fig)
```
