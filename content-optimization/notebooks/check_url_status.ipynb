### Function to check if url exists

```python
def check_url_exists(url):
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36"
    }
    try:
        # Get response of URL
        response = requests.get(url, headers=headers, allow_redirects=True, timeout=10)
        # Get page content to check if it contains 404 error
        page_content = response.text.lower()
        # Constant variable for error status code
        errorStatusCode = 400

        # Check if page contains "404 error" in its content
        if "404 error" in page_content:
            status = "404 error"
            return False, status
        # Check if final URL is different from original url
        elif response.url != url:
            status = f"Redirected to '{response.url}'."
            return False, status
        # If response is not a success code, treat it as not existing
        elif response.status_code >= errorStatusCode:
            status = "400 error"  # Bad request
            return False, status
        # If no exception is raised, URL exists
        else:
            status = "URL exists"
            return True, status
    except requests.RequestException as e:
        return False, str(e)
```

### Main Program

```python
# Takes ~18 minutes to run
# Store URLs' status
url_status = []

# Loop through merged_dataset to check validity of URLs
for i, row in df.iterrows():
    url = row["full_url"]
    result, status_message = check_url_exists(url)
    if result:
        print(f"{i+1}. ", status_message)
    else:
        print(f"{i+1}. ", status_message)

    url_status.append(status_message)

# Save to txt file to avoid re-running again in the future
with open("url_check.txt", "w") as output:
    output.write(str(url_status))

# Create a new column in the DataFrame with the collected status messages
df["url_check"] = url_status
```

```python
# Clean `extracted_content_body`


def clean_illegal_characters_in_column(df, column_name):
    # Define a pattern for illegal characters (ASCII control characters)
    illegal_characters_pattern = r"[\x00-\x1F\x7F-\x9F]"  # ASCII control characters

    # Clean specified column
    df[column_name] = df[column_name].apply(
        lambda x: re.sub(illegal_characters_pattern, "", x) if isinstance(x, str) else x
    )

    return df


df_copy = clean_illegal_characters_in_column(df_copy, "extracted_content_body")
```
