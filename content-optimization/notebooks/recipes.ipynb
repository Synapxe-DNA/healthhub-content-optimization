## `title`


##


```python
recipe_title_indices = np.argwhere(
    [
        True if title is not None and re.search(r"[rR]ecipes?", title) else False
        for title in filtered_data["title"]
    ]
).flatten()

title_recipes = filtered_data.iloc[recipe_title_indices].id.to_list()
```

## `keywords`


```python
recipe_kw_indices = np.argwhere(
    [
        True if kw is not None and re.search(r"[rR]ecipes?", kw) else False
        for kw in filtered_data["keywords"]
    ]
).flatten()

kw_recipes = filtered_data.iloc[recipe_kw_indices].id.to_list()
```

## `extracted_content_body`


```python
recipe_content_indices = np.argwhere(
    [
        (
            True
            if content is not None
            and re.search(r"what [do ]?you need", content.lower())
            and re.search(r"how to cook [this dish]*", content.lower())
            else False
        )
        for content in filtered_data["extracted_content_body"]
    ]
).flatten()

content_recipes = filtered_data.iloc[recipe_content_indices].id.to_list()
```

## `title` + `keywords` + `extracted_content_body`


```python
indices = set(title_recipes).union(set(kw_recipes)).union(set(content_recipes))

recipes_data = filtered_data.query("id in @indices")
recipes_data
```

```python
recipes_data = recipes_data[
    [
        "id",
        "content_category",
        "pr_name",
        "content_name",
        "title",
        "full_url",
        "content_body",
        "extracted_content_body",
        "to_remove",
        "remove_type",
    ]
]
```
