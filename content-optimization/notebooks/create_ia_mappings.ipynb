```python
merged_data = catalog.load("merged_data")

display(merged_data)
```

```python
df = merged_data[
    [
        "id",
        "content_name",
        "title",
        "article_category_names",
        "full_url",
        "content_body",
        "extracted_content_body",
        "pr_name",
        "date_modified",
        "content_category",
    ]
]
display(df)
```

```python
unique_names = df["article_category_names"].unique()

print(f"Total unique names: {len(unique_names)}")
print(f"Unique names: {unique_names}")
```

```python
list_of_names = list(unique_names)
print(f"List of unique names: {list_of_names}")

cat_names = []
for cat in list_of_names:
    if isinstance(cat, str):
        categories = cat.split(",")
        for category in categories:
            amended_category = category.replace("&", "and")
            cat_names.append(amended_category)
    else:
        cat_names.append(cat)
```

```python
cat_names = list(
    set(filter(lambda x: len(x) > 0 if x is not None else True, cat_names))
)

print(f"Total unique names: {len(cat_names)}")
print("List of unique names:")
for name in cat_names:
    print(name)
```

```python
l1_mappings = catalog.load("params:l1_mappings")
print(l1_mappings)
```

```python
def invert_mappings(mappings):
    invert_mappings = {}
    for content_category, values in mappings.items():
        map_dicts = {}
        for l1_map, category_names in values.items():
            for category_name in category_names:
                map_dicts[category_name] = l1_map

        invert_mappings[content_category] = map_dicts
    return invert_mappings


print(invert_mappings(l1_mappings))
```

```python
l2_mappings = catalog.load("params:l2_mappings")
print(l2_mappings)
```

```python
print(invert_mappings(l2_mappings))
```

```python
def map_category_names(
    mappings, df, content_category_column, reference_column, new_column_name
):
    df[new_column_name] = None
    for index, row in df.iterrows():
        category_string = row[reference_column]
        content_category = row[content_category_column]
        category_map = mappings.get(content_category, None)

        if category_string is None:
            continue

        category_string = category_string.replace("&", "and")
        print(f"article_category_names: {category_string}")
        df.at[index, reference_column] = category_string
        categories_list = list(filter(lambda x: len(x) > 0, category_string.split(",")))

        if category_map is not None:
            for ind, category_name in enumerate(categories_list):
                result = category_map.get(category_name, category_name)
                print(
                    f"Content category: {content_category}, Article Category: {category_name}, New Mapping: {result}"
                )
                categories_list[ind] = result

            df.at[index, new_column_name] = " | ".join(list(set(categories_list)))
    return df


df = map_category_names(
    invert_mappings(l1_mappings),
    df,
    "content_category",
    "article_category_names",
    "l1_mappings",
)
display(df)
```

```python
df = map_category_names(
    invert_mappings(l2_mappings),
    df,
    "content_category",
    "article_category_names",
    "l2_mappings",
)
display(df)
```
